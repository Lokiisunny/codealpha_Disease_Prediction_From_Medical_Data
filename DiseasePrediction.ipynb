importing required file

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
Data collection and Analysis Disease prediction Data set
```

```python
data_set = pd.read_csv("medical_disease_multiclass.csv")
```

```python
# printing the first five rows
data_set.head()
```

```python
# number of rews and columns in this dataset
data_set.shape
```

```python
# getting the statistical measures of the data
data_set.describe()
```

```python
data_set['disease'].value_counts()
```

```python
data_set.groupby('disease').mean()
```

```python
# separating data set labels
X = data_set.drop(columns = 'disease', axis = 1)
Y = data_set['disease']
```

```python
# Encode target labels
le = LabelEncoder()
y_encoded = le.fit_transform(Y)
```

```python
print(X)
print(y_encoded)
```

Train Test Split data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape, y_test.shape, y_train.shape)
```

Training Model

```python
model = LogisticRegression(max_iter=1000, multi_class='multinomial', solver='lbfgs')
model.fit(X_train, y_train)
```

Model Evaluation

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
// Using Support vector Machine Model 
```

```python
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale')
svm_model.fit(X_train, y_train)
```

```python
# Predictions and evaluation
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("SVM Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
using Random Forest
```

```python
# Train Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Predictions and evaluation
y_pred = rf_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Random Forest Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```
